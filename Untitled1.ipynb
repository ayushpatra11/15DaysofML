```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=df.drop(['attack_cat','proto','service','state','dur','spkts','dpkts','sbytes','dbytes','rate','sttl','dttl','sload','dload','sloss','dloss','sinpkt','dinpkt','sjit','djit','swin','stcpb'], axis=1)
```

```python
df.columns
```

```python
def plotbox(s):
    sns.boxplot(data=df, y=s, x='label', showfliers=False)
    plt.show()
```

```python
col=df.columns
for i in range(len(col)):
    plotbox(col[i])
```

```python
def plotdist(s):
    sns.distplot(df[df["label"]==0][s], hist=False,label="No Attack")
    sns.distplot(df[df["label"]==1][s], hist=False,label="Attack")
    plt.legend()
    plt.show()
```

```python
norm=[]
for i in range(len(col)-2):
    try:
        plotdist(col[i])
    except:
        norm.append(col[i])
```
